```python
# Prepare a Naive Bayes classification model for predicting the purchase power of user
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import precision_recall_curve
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
```

```python
df = pd.read_csv("User_Data.csv")
```

```python
df.head()
```

```python
df.drop(columns = ['User ID'],axis = 1, inplace = True)
```

```python
df.info()
```

```python
# Label Encoding (Also known as One hot Encoding)
le = LabelEncoder()
df['Gender'] = le.fit_transform(df['Gender'])

# Label Encoder is a class which is used to convert a categorical variable into numerical variable
# Since a ML model is a mathematical model, so it understands only numerical values
```

```python
# Split data into dependant/independent variable
x = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
# Split data into test/train set
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.25, random_state = True)
```

```python
# Scale dataset
sc = StandardScaler()
X_train = sc.fit_transform(x_train)
X_test = sc.transform(x_test)
```

```python
# Classifier
classifier = GaussianNB()
classifier.fit(X_train, y_train)
```

```python
# Prediction
y_pred = classifier.predict(X_test)
# Accuracy
accuracy_score(y_test, y_pred)
```

```python
# Classification report
print(f'Classification Report : \n {classification_report(y_test, y_pred)}')
```

```python
# Confusion matrix
cf_matrix = confusion_matrix(y_test, y_pred)
print(cf_matrix)
```
